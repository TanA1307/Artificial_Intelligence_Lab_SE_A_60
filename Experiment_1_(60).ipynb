```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

# Example usage:
number = 5
num = int(input("Enter a non negative integer: "))
print(f"The factorial of {num} is {factorial(num)}")

```

```python
# Function to display the multiplication table
def display_multiplication_table(num):
    print(f"Multiplication Table for {num}:")
    for i in range(1, 11): # Table from 1 to 10
        print(f"{num} x {i} = {num * i}")

# Get input from the user
number_for_table = int(input("Enter a number to display its multiplication table: "))

# Call the function to display the table
display_multiplication_table(number_for_table)
```

```python
# Get input from the user
num = int(input("Enter an integer: "))

# Check if the number is odd or even
if (num % 2) == 0:
    print(f"The number {num} is even.")
else:
    print(f"The number {num} is odd.")
```

```python
# Get input from the user
num_str = input("Enter a number: ")

# Reverse the string
reversed_num_str = num_str[::-1]

# Check if it's a palindrome
if num_str == reversed_num_str:
    print(f"The number {num_str} is a palindrome.")
else:
    print(f"The number {num_str} is not a palindrome.")
```

```python
class ExpertSystem:
    def decide(self, weather):
        if weather == "rainy":
            return "Stay inside"
        elif weather == "sunny":
            return "Go outside"
        else:
            return "No decision"


# Use the expert system
system = ExpertSystem()

weather = input("Enter weather (sunny/rainy): ")
decision = system.decide(weather)

print("Decision:", decision)
```
