```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given parameters
mu = 0        # Mean
sigma = 1     # Standard Deviation

# Create an array of x values
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# Calculate the PDF using scipy
pdf = norm.pdf(x, mu, sigma)

# Plot the normal distribution
plt.figure()
plt.plot(x, pdf)
plt.title("Normal Distribution (μ = {}, σ = {})".format(mu, sigma))
plt.xlabel("X values")
plt.ylabel("Probability Density")
plt.grid()
plt.show()
```

```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Sample dataset
data = np.array([12, 15, 14, 10, 18, 20, 17, 16, 19, 11])

# Descriptive Statistics
mean = np.mean(data)
median = np.median(data)
variance = np.var(data)
std_deviation = np.std(data)

# Using SciPy (for verification of mean)
mean_scipy = stats.tmean(data)

# Print results
print("Dataset:", data)
print("Mean (μ):", mean)
print("Median:", median)
print("Variance (σ²):", variance)
print("Standard Deviation (σ):", std_deviation)

# Plotting the data
plt.figure()
plt.hist(data, bins=5)
plt.axvline(mean, linestyle='dashed')
plt.axvline(median, linestyle='dotted')

plt.title("Descriptive Statistics Visualization")
plt.xlabel("Data Values")
plt.ylabel("Frequency")
plt.show()

```

```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Number of coin tosses
n_tosses = 1000

# Simulate coin tosses (0 = Tail, 1 = Head)
tosses = np.random.choice([0, 1], size=n_tosses)

# Count number of Heads and Tails
heads = np.sum(tosses)
tails = n_tosses - heads

# Empirical probabilities
p_heads = heads / n_tosses
p_tails = tails / n_tosses

# Theoretical probabilities using SciPy (Binomial distribution)
theoretical_prob = stats.binom.pmf(heads, n_tosses, 0.5)

# Print results
print("Total Tosses:", n_tosses)
print("Number of Heads:", heads)
print("Number of Tails:", tails)
print("Empirical P(Head):", p_heads)
print("Empirical P(Tail):", p_tails)

# Plot results
labels = ['Head', 'Tail']
counts = [heads, tails]

plt.figure()
plt.bar(labels, counts)
plt.title("Coin Toss Simulation (n = {})".format(n_tosses))
plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.show()

```
